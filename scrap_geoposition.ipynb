```python
headers = {
"User-Agent": "city_timezone_converter"
}


# new_column_lat = []
# new_column_lon = []
for index in range(df_hotels.shape[0]):
    if new_column_lat[index]==None:
        if df_hotels['Адрес'][index] not in ['Подробнее', 'Позвонить', 'Посетить веб-сайт', 'Эл. почта', np.nan]:
            print(index, df_hotels['Адрес'][index])
            url = "https://nominatim.openstreetmap.org/search?q="+df_hotels['Адрес'][index].replace(",", "%2C").replace(" ", "+")+"&format=json"
        # print(url)

            time.sleep(random.randint(5, 10)) # Задержка между запросами

            response = requests.get(url, headers=headers)

            if response.status_code == 200:
                # print(response.json())
                if response.json()!=[]:
                    lat = response.json()[0]['lat'] # Широта
                    lon = response.json()[0]['lon'] # Долгота
                else:
                    lat, lon = None, None
            
                new_column_lat[index] = lat
                new_column_lon[index]= lon
            # print(new_column_lat)
            else:
                print(f"Error: {response.status_code}")

 
# # Добавляем новый столбец в DataFrame
# df_hotels['Широта'] = new_column_lat
# df_hotels['Долгота'] = new_column_lon
```
