```python
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
# Загрузка данных из CSV файлов
hotels_df = pd.read_csv('df_hotels_toBD.csv').drop(columns=['Unnamed: 0'])
dostoprim_df = pd.read_csv('df_dostoprim_toBD.csv').drop(columns=['Unnamed: 0'])
styles_df = pd.read_csv('df_styles_toBD.csv').drop(columns=['Unnamed: 0'])
# hotels_df.head()
# dostoprim_df.head()
```

```python
# Настройка подключения к базе данных PostgreSQL
DATABASE_TYPE = 'postgresql'
DBAPI = 'psycopg2'
HOST = 'localhost'
USER = 'postgres'  # замените на ваше имя пользователя
PASSWORD = '12345'  # замените на ваш пароль
DATABASE = 'DB_hotels'  # замените на имя вашей базы данных

# Создание строки подключения
connection_string = f"{DATABASE_TYPE}+{DBAPI}://{USER}:{PASSWORD}@{HOST}/{DATABASE}"
engine = create_engine(connection_string)
```

```python
# Загрузка данных о отелях в базу данных
hotels_df.to_sql('hotels_info', engine, if_exists='replace', index=False)

# Загрузка данных о достопримечательностях в базу данных
dostoprim_df.to_sql('dostoprim_info', engine, if_exists='replace', index=False)

# Загрузка данных о стилях отеля в базу данных
dostoprim_df.to_sql('dostoprim_info', engine, if_exists='replace', index=False)
```

```python
with engine.connect() as connection:
    result = connection.execute(text("SELECT * FROM hotels_info LIMIT 5;"))
    for row in result:
        print(row)

    result = connection.execute(text("SELECT * FROM dostoprim_info LIMIT 5;"))
    for row in result:
        print(row)

    result = connection.execute(text("SELECT * FROM styles_info LIMIT 5;"))
    
```
